```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from sklearn.metrics import r2_score
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_curve, roc_auc_score, plot_roc_curve
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score, plot_roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
x = df[['STATE/UT_encoded','TOTAL IPC CRIMES']]
x
```

#### Applying the K-means algorithm to a dataset

```python
kmeans = KMeans(n_clusters = 3,random_state = 70)
y_kmeans = kmeans.fit_predict(x)
print(y_kmeans)
```

MODEL EVALUATION

```python
silhouette_score(x,y_kmeans)
```

### 4.2 
Partition states according to clusters.

```python
sensitive_cluster = df[y_kmeans==0]
print(sensitive_cluster)
print('\n')
moderate_cluster = df[y_kmeans==2]
print(moderate_cluster)
peaceful_cluster = df[y_kmeans==1]
print(peaceful_cluster)
print('\n')
```

Creating a visual representation of the clusters.

```python
plt.figure(figsize = (8,6))
plt.scatter(x[y_kmeans==0]['STATE/UT_encoded'],x[y_kmeans==0]['TOTAL IPC CRIMES'],s=100, c = 'green', label = 'Sensitive Areas')
plt.scatter(x[y_kmeans==1]['STATE/UT_encoded'],x[y_kmeans==1]['TOTAL IPC CRIMES'],s=100, c = 'cyan', label = 'Peaceful Areas')
plt.scatter(x[y_kmeans==2]['STATE/UT_encoded'],x[y_kmeans==2]['TOTAL IPC CRIMES'],s=100, c = 'black', label = 'Moderate Areas')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.title('Clusters of States/UTs')
plt.xlabel('States/UTs')
plt.ylabel('Total Crime')
plt.legend()
plt.show()
```

Illustrating the clusters in accordance with the states.

```python
plt.figure(figsize = (50,10))
plt.scatter(df[y_kmeans==0]['STATE/UT'],df[y_kmeans==0]['TOTAL IPC CRIMES'],s=100, c = 'red', label = 'Sensitive Areas')
plt.scatter(df[y_kmeans==1]['STATE/UT'],df[y_kmeans==1]['TOTAL IPC CRIMES'],s=100, c = 'cyan', label = 'Peaceful Areas')
plt.scatter(df[y_kmeans==2]['STATE/UT'],df[y_kmeans==2]['TOTAL IPC CRIMES'],s=100, c = 'magenta', label = 'Moderate Areas')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.title('Clusters of States/UTs')
plt.xlabel('States/UTs')
plt.ylabel('Total Crime')
plt.legend()
plt.show()
```

### PROJECT REPORT

##### Project Overview:
This report examines crime data for different states and union territories in India. It aims to understand why some places have more crime than others. We look at two things: how sensitive a place is to crime and how peaceful it is. We also study if there's a connection between crime and things like how many people live there, how educated they are, and how big the place is.

##### How We Studied:
We used data from the years 2001 and 2012, which includes information about different crimes, the number of people, literacy rates, and the size of each place. We found which places had the most sensitive and peaceful situations based on this data.

##### What We Found:

##### Sensitive States:
In 2001, the states with the most sensitivity to crime were Madhya Pradesh, Maharashtra, Rajasthan, Tamil Nadu, and Uttar Pradesh. They had higher crime rates for things like murder, rape, and theft.

##### Peaceful States:
In 2001, the most peaceful states were Nagaland, Sikkim, Andaman and Nicobar Islands, Daman and Diu, Dadra and Nagar Haveli, and Lakshadweep. They had lower crime rates.

##### Literacy and Crime:
We noticed that places with higher literacy rates, like Kerala, Mizoram, Tripura, Goa, and Himachal Pradesh, tended to have lower crime rates. In contrast, states with lower literacy rates, such as Bihar, Arunachal Pradesh, Rajasthan, Uttar Pradesh, and Jharkhand, had higher crime rates.

##### Types of Crime and Literacy:
Different places showed different patterns for types of crimes and literacy rates. Some states with lower crime rates and higher literacy rates were Mizoram, Sikkim, Nagaland, Tripura, and Manipur. Meanwhile, Delhi, Assam, Bihar, Rajasthan, and Uttar Pradesh had higher crime rates and varying literacy levels.

##### Change Over Time:
When we looked at the crime rates from 2001 to 2012, we saw that crime generally increased over the years. However, we need to study each place more closely to understand specific trends.

##### Size of Area and Crime:
Big states like Rajasthan, Madhya Pradesh, Maharashtra, and Uttar Pradesh had higher crime rates. But remember, the size of the area alone doesn't determine crime rates. Other things like how rich or poor people are, how many police there are, and the local culture also matter.

##### Population and Crime:
States with more people, like Uttar Pradesh, Maharashtra, Bihar, West Bengal, and Andhra Pradesh, tended to have more crime. But just having a lot of people isn't the only reason. Other factors like how people live together, their jobs, and how well the police work also make a difference.

##### In Conclusion:
After looking at the data, we can say that having more educated people can help reduce crime. But it's not just about that – bigger states and more people can also lead to more crime. Many things affect crime rates, including how rich or poor people are and how well the police do their job.
